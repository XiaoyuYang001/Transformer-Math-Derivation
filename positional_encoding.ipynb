```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def sinusoidal_pos_encoding(seq_len, d_model):
    position = np.arange(seq_len)[:, np.newaxis]  # (seq_len, 1)
    div_term = np.exp(np.arange(0, d_model, 2) * (-np.log(10000.0) / d_model))  # 修正括号
    pe = np.zeros((seq_len, d_model))
    pe[:, 0::2] = np.sin(position * div_term)  # 偶数位置用sin
    pe[:, 1::2] = np.cos(position * div_term)  # 奇数位置用cos
    return pe

# 参数设置
seq_len = 50
d_model = 512
pe = sinusoidal_pos_encoding(seq_len, d_model)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(pe[:30, :64], cmap="viridis")  # 只显示前30个位置和前64维
plt.title("Sinusoidal Positional Encoding (First 64 Dimensions)")
plt.xlabel("Dimension")
plt.ylabel("Position")
plt.savefig("figs/sinusoidal_pe.png", dpi=300, bbox_inches='tight')
plt.show()
```
